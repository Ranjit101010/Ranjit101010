```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde = False) # univariate variables
```

```python
sns.histplot(tips['total_bill'],bins = 20,kde= True)
```

```python
tips['total_bill'].describe()
```

```python
sns.histplot(tips['total_bill'])
```

```python
# bivariate data
sns.jointplot(x = 'total_bill',y = 'tip',data = tips) # by default it gives scatter 
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex')
```

```python
sns.pairplot(tips,hue = 'sex',palette = 'coolwarm')
```

# categorical dataplots

```python
# Bar plot

sns.barplot(x = 'sex',y = 'total_bill',data = tips)

'x column should be a categorical data'
'y column should be a numerical data'
```

```python
import numpy as np
```

```python
sns.barplot(x = 'sex',y = 'total_bill',data = tips,estimator = np.std)
```

```python
sns.barplot(x = 'sex',y = 'total_bill',data = tips,estimator = np.mean)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,10))
ax = sns.barplot(x = 'sex',y = 'total_bill',hue = 'day',data = tips,estimator = np.median)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = tips,hue = 'smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue = 'sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue = 'sex',split = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,dodge=False,hue='sex')
```
